```python
import math as h
import statistics as st

def datos():
    dat=list()
    n=int(input("Cantidad de datos a ingresar: "))
    for i in range(n):
        p=float(input("Ingresa el dato: "))
        dat.append(p)
    return dat

def ingr():
    pd=int(input("Se tiene\n1: Muestra de datos\n2: Suma de datos\n\nRespuesta: "))
    while pd<1 or pd>2:
        pd=int(input("\nOpción inválida. Respuesta: "))
    if pd==1:
        e=list()
        e=datos()
        nm=len(e)
        sm=sum(e)
        vrz=st.variance(e)
    if pd==2:
        sm=float(input("Ingrese la suma de los datos: "))
        nm=int(input("Ingrese la cantidad de la muestra: "))
        vrz=float(input("Ingresa la varianza de los datos: "))
    np=int(input("Ingresa la población total: "))
    return nm,np,sm,vrz

def inp():
    ex=float(input("Ingresa el número de éxitos: "))
    nm=int(input("Ingrese la cantidad de la muestra: "))
    np=int(input("Ingresa la población total: "))
    p=ex/nm
    return p,nm,np

def mmas(nm,np,sd,vrm):
    xb=sd/nm
    vr=(vrm/nm)*((np-nm)/np)
    b=2*(h.sqrt(vr))
    li=xb-b
    ls=xb+b
    me=(b/xb)*100
    print("\n\n\nLa media poblacional es:",xb)
    print("Con una varianza de: ",vr)
    print("Una cota para el error de estimación de: ",b," y un margen de error del: ",me,"%.")
    print("Con una confianza del 95%, la media poblacional se encuentra en el intevalo de confianza: ",li," < M < ",ls,"\n\n")

def tmas(nm,np,sd,vrm):
    xb=sd/nm
    t=xb*np
    vr=((vrm*np)/nm)*(np-nm)
    b=2*(h.sqrt(vr))
    li=t-b
    ls=t+b
    me=(b/t)*100
    print("\n\n\nEl total poblacional es:",t)
    print("Con una varianza de: ",vr)
    print("Una cota para el error de estimación de: ",b," y un margen de error del: ",me,"%.")
    print("Con una confianza del 95%, el total poblacional se encuentra en el intevalo de confianza: ",li," < t < ",ls,"\n\n")

def pmas(p,nm,np):
    sigma=p*(1-p)
    vr=(sigma/nm)*((np-nm)/(np-1))
    b=2*(h.sqrt(vr))
    li=p-b
    ls=p+b
    me=(b/p)*100
    print("\n\n\nLa proporción poblacional es:",p)
    print("Con una varianza de: ",vr)
    print("Una cota para el error de estimación de: ",b," y un margen de error del: ",me,"%.")
    print("Con una confianza del 95%, la proporción poblacional se encuentra en el intevalo de confianza: ",li," < p < ",ls,"\n\n")

def tm1a():
    tom=int(input("1. Media o Proporción\n2. Total\n\nRespuesta: "))
    b=float(input("Error máximo: "))
    np=int(input("Ingresa la población total: "))
    if tom==1:
        d=(b*b)/4
    if tom==2:
        d=((b*b)/(4*np*np))
    vrz=float(input("Ingresa la varianza: "))
    nap=(np*vrz)/(((np-1)*d)+vrz)
    print("\n\nLa población necesaria es: ",nap)

def cost(l):
    ci=list()
    for i in range(l):
        print("\nEstrato ",i+1,": ")
        c=float(input("Ingrese el costo del estrato: "))
        ci.append(c)
    return ci

def conest(np,ni,l,vr):
    s=list()
    wi=list()
    op2=int(input("\n\n1: Estratos iguales\n2: Costos iguales\n3. Misma volatilidad\n4. Asignación proporcional\n5: Asignación equitativa\n\nRespuesta: "))
    while op2<1 or op2>5:
        op2=int(input("\nOpción inválida. Respuesta: "))
    if op2==1:
        ci=cost(l)
        for j in range(l):
            h=vr[j]/ci[j]
            s.append(h)
        for j in range(l):
            w=(vr[j]/ci[j])/sum(s)
            wi.append(w)
    if op2==2:
        for j in range(l):
            h=vr[j]*ni[j]
            s.append(h)
        for j in range(l):
            w=(vr[j]*ni[j])/sum(s)
            wi.append(w)
    if op2==3:
        ci=cost(l)
        for j in range(l):
            h=ni[j]/ci[j]
            s.append(h)
        for j in range(l):
            w=(ni[j]/ci[j])/sum(s)
            wi.append(w)
    if op2==4:
        for j in range(l):
            w=ni[j]/np
            wi.append(w)
    if op2==5:
        for j in range(l):
            w=1/l
            wi.append(w)
    return wi

def estr():
    sm=list()
    ni=list()
    vr=list()
    pe=list()
    ne=int(input("Ingresa el número de estratos: "))
    for j in range(ne):
        print("\nEstrato ",j+1,":")
        d=ingr()
        pe.append(d[1])
        x=d[2]/d[0]
        sm.append(x)
        ni.append(d[0])
        vr.append(d[3])
    np=sum(pe)
    wi=conest(np,pe,ne,vr)
    return wi,sm,ni,vr,ne,np,pe

def pet():
    pi=list()
    ni=list()
    vr=list()
    ex=list()
    pe=list()
    ne=int(input("Ingresa el número de estratos: "))
    for j in range(ne):
        print("\nEstrato ",j+1,":")
        d=inp() #pr,muestra,pob
        pe.append(d[2])
        ni.append(d[1])
        v=d[0]*d[1]
        ex.append(v)
    np=sum(pe)
    if sum(ex)!=sum(ni):
        for a in range(ne):
            p=ex[a]/ni[a]
            x=p*(1-p)
            pi.append(p)
            vr.append(x)
    else:
        for a in range(ne):
            p=ex[a]/sum(ni)
            x=p*(1-p)
            pi.append(p)
            vr.append(x)
    wi=conest(np,pe,ne,vr)
    return wi,pi,ni,vr,ne,np,pe

def mmae(wi,sm,ni,vr,ne,np,pe):
    s=list()
    sv=list()
    for j in range(ne):
        g=pe[j]*sm[j]
        s.append(g)
    xb=sum(s)/np
    for j in range(ne):
        r=(vr[j]*pe[j]*(pe[j]-ni[j]))/ni[j]
        sv.append(r)
    vr=sum(sv)/(np*np)
    b=2*(h.sqrt(vr))
    li=xb-b
    ls=xb+b
    me=(b/xb)*100
    print("\n\n\nLa media poblacional es:",xb)
    print("Con una varianza de: ",vr)
    print("Una cota para el error de estimación de: ",b," y un margen de error del: ",me,"%.")
    print("Con una confianza del 95%, la media poblacional se encuentra en el intevalo de confianza: ",li," < M < ",ls,"\n\n")
    
def tmae(wi,sm,ni,vr,ne,np,pe):
    s=list()
    sv=list()
    for j in range(ne):
        g=pe[j]*sm[j]
        s.append(g)
    t=sum(s)
    for j in range(ne):
        r=(vr[j]*pe[j]*(pe[j]-ni[j]))/ni[j]
        sv.append(r)
    vr=sum(sv)
    b=2*(h.sqrt(vr))
    li=t-b
    ls=t+b
    me=(b/t)*100
    print("\n\n\nEl total poblacional es:",t)
    print("Con una varianza de: ",vr)
    print("Una cota para el error de estimación de: ",b," y un margen de error del: ",me,"%.")
    print("Con una confianza del 95%, el total poblacional se encuentra en el intevalo de confianza: ",li," < t < ",ls,"\n\n") 

def pmae(wi,pi,ni,vr,ne,np,pe):
    s=list()
    sv=list()
    for j in range(ne):
        g=pe[j]*pi[j]
        s.append(g)
    p=sum(s)/np
    for j in range(ne):
        r=(vr[j]*pe[j]*(pe[j]-ni[j]))/(ni[j]-1)
        sv.append(r)
    vr=sum(sv)/(np*np)
    b=2*(h.sqrt(vr))
    li=p-b
    ls=p+b
    me=(b/p)*100
    print("\n\n\nLa proporción poblacional es:",p)
    print("Con una varianza de: ",vr)
    print("Una cota para el error de estimación de: ",b," y un margen de error del: ",me,"%.")
    print("Con una confianza del 95%, la proporción poblacional se encuentra en el intevalo de confianza: ",li," < p < ",ls,"\n\n") 

def tmue():
    ni=list()
    vi=list()
    wi=list()
    tom=int(input("1. Media o Proporción\n2. Total\n\nRespuesta: "))
    b=float(input("Error máximo: "))
    np=int(input("Ingresa la población total: "))
    if tom==1:
        d=(b*b)/4
    if tom==2:
        d=((b*b)/(4*np*np))
    l=int(input("Ingresa el número de estratos que se tienen: "))
    for j in range(l):
        print("Estrato ",j+1,":")
        n=int(input("Ingresa la población del estrato: "))
        w=float(input("Ingresa el peso del estrato: "))
        v=float(input("Ingresa la varianza del estrato: "))
        ni.append(n)
        vi.append(v)
        wi.append(w)
    ar=list()
    ab=list()
    for j in range(l):
        q=(ni[j]*ni[j]*vi[j])/wi[j]
        f=ni[j]*vi[j]
        ar.append(q)
        ab.append(f)
    nap=sum(ar)/((d*np*np)+sum(ab))
    ppe=list()
    for j in range(l):
        y=wi[j]*nap
        ppe.append(y)
    print("\n\nLa población necesaria es: ",nap)
    for j in range(l):
        print("El estrato ",j+1," debe tener una población de: ",ppe[j])

op=0
while op!=3:
    op=int(input("1: Muestreo Aleatorio Simple\n2: Muestreo Aleatorio Estratificado\n3: Salir\n\nRespuesta: "))
    while op<1 or op>3:
        op=int(input("\nOpción inválida. Respuesta: "))
    if op==1:
        op2=int(input("1: Media Poblacional\n2: Total Poblacional\n3. Proporción Poblacional\n4. Tamaño de Muestra\n5: Salir\n\nRespuesta: "))
        while op2<1 or op2>5:
            op2=int(input("\nOpción inválida. Respuesta: "))
        if op2==1:
            ba=ingr()
            mmas(ba[0],ba[1],ba[2],ba[3])
        if op2==2:
            ba=ingr()
            tmas(ba[0],ba[1],ba[2],ba[3])
        if op2==3:
            ba=inp()
            pmas(ba[0],ba[1],ba[2])
        if op2==4:
            tm1a()
    if op==2:
        op2=int(input("1: Media Poblacional\n2: Total Poblacional\n3. Proporción Poblacional\n4. Tamaño de Muestra\n5: Asignación de un estrato\n6. Salir\n\nRespuesta: "))
        while op2<1 or op2>6:
            op2=int(input("\nOpción inválida. Respuesta: "))
        if op2==1:
            p=estr()
            print("\nLos estratos son: ",p[0],"\n")
            mmae(p[0],p[1],p[2],p[3],p[4],p[5],p[6])
        if op2==2:
            p=estr()
            print("\nLos estratos son: ",p[0],"\n")
            tmae(p[0],p[1],p[2],p[3],p[4],p[5],p[6])
        if op2==3:
            k=pet()
            pmae(k[0],k[1],k[2],k[3],k[4],k[5],k[6])
        if op2==4:
            tmue()
        if op2==5:
            p=estr()
            print("Los estratos son: ",p[0],"\n\n")
```
